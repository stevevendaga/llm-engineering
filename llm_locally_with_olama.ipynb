```python
#imports
import requests
from openai import OpenAI
```

```python

# Check if Ollama is running by making a proper request
try:
    response = requests.get("http://localhost:11434")
    print(f"Status Code: {response.status_code}")
    print(f"Response: {response.text}")
except requests.exceptions.ConnectionError:
    print("Cannot connect to Ollama. Make sure it's running on localhost:11434")
except Exception as e:
    print(f"An error occurred: {e}")
```

```python
# Install llama3.2 model from ollama
# !ollama pull llama3.2
# !ollama pull gpt-oss:20b
```

```python
# Declarations
OLLAMA_BASE_URL  = "http://localhost:11434/v1"

ollama = OpenAI(base_url=OLLAMA_BASE_URL, api_key="ollama")

olama_model = "llama3.2"
```

```python
# Message prompt
user_message = "Where is makurdi located in nigeria"
```

```python


response = ollama.chat.completions.create(model=olama_model,
                                          messages=[{"role": "user", "content":user_message}])
print(response.choices[0].message.content)
```
